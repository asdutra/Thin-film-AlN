```python
import matplotlib.pyplot as plt

# Leitura do arquivo e processamento dos dados
results = []
with open('AlN300nm_532_1200_25pc_10sx5acc.txt', 'r') as f:
    for i, l in enumerate(f):
        if i > 36:
            results = [[float(entry) for entry in line.split()] for line in f.readlines()]
            #print(results)

# Separando os dados em listas de x e y
x_values = [row[0] for row in results]
y_values = [row[1] for row in results]

# Plotando os dados
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, color='b', label='AlN300nm_532_1200_25pc_10sx5acc')
plt.ylabel('Intensity (Counts)')
plt.xlabel('Raman shift ( Cm-1')
plt.title('Espectro raman AlN 300nm, grade de 1200, 25% de potncia de laser, 10 segundos por contagem, 5 contagens')
plt.legend()
plt.grid()
plt.savefig("AlN300nm_532_1200_25pc_10sx5acc.jpg")
plt.show()

```
