```python
#Em função:
import matplotlib.pyplot as plt
import os
import numpy as np

lista_de_plots = []

def plotar_espectro(caminho_pasta, nome_arquivo_entrada, nome_arquivo_saida):
    # Caminho do arquivo de entrada
    caminho_arquivo = os.path.join(caminho_pasta, nome_arquivo_entrada)

    # Leitura do arquivo e processamento dos dados
    results = []
    with open(caminho_arquivo, 'r') as f:
        for i, l in enumerate(f):
            if i > 36:  # Ignora as primeiras 36 linhas
                # Adiciona as linhas restantes na lista 'results'
                results.append([float(entry) for entry in l.split()])

    # Separando os dados em listas de x e y
    x_values = [row[0] for row in results]
    y_values = [row[1] for row in results]
    y_values = np.array(y_values)

    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    y_values = scaler.fit_transform(y_values.reshape(-1, 1)).flatten()
    lista_de_plots.append(x_values)
    lista_de_plots.append(y_values)

    # Plotando os dados
    plt.figure(figsize=(12, 8))
    plt.plot(x_values, y_values, color='b', label='Aquisição superficial da amostra', linewidth=0.5)
    plt.ylabel('Intensity (Counts)')
    plt.xlabel('Raman shift (Cm-1)')
    legenda = input("Escreva a legenda do plot")
    plt.title(legenda)
    plt.legend()
    plt.grid()

    # Caminho de salvamento do arquivo de saída
    caminho_saida = os.path.join(caminho_pasta, nome_arquivo_saida)
    plt.savefig(caminho_saida)

    # Exibe o gráfico
    plt.show()




# Chamando para gráfico com potencia baixa de laser (AlN + Sic)
pasta = r"C:\Users\Agatha\CÓPIA - Programas.Thin.Film.AlN\ESPECTROMETRIA-12fev2025"
nome_arquivo_entrada = "AlN+SiC_532_1800_01pc_40sx5acc-150-1750.txt"
nome_arquivo_saida = "AlN+SiC_532_1800_01pc_40sx5acc-150-1750.jpg"
plotar_espectro(pasta, nome_arquivo_entrada, nome_arquivo_saida)
print(len(lista_de_plots))
#Somente sic em alta potencia 
pasta = r"C:\Users\Agatha\CÓPIA - Programas.Thin.Film.AlN\ESPECTROMETRIA-12fev2025"
nome_arquivo_entrada = "SiC_532_1800_50pc_10sx5acc-150-1750.txt"
nome_arquivo_saida = "SiC_532_1800_50pc_10sx5acc-150-1750.jpg"
plotar_espectro(pasta, nome_arquivo_entrada, nome_arquivo_saida)
print((lista_de_plots[1]))
print((lista_de_plots[3]))

# AlN + Sic em alta potência 
pasta = r"C:\Users\Agatha\CÓPIA - Programas.Thin.Film.AlN\ESPECTROMETRIA-12fev2025"
nome_arquivo_entrada = "AlN150nm_532_1800_50pc_10sx5acc-150-1750.txt"
nome_arquivo_saida = "AlN150nm_532_1800_50pc_10sx5acc-150-1750.jpg"
plotar_espectro(pasta, nome_arquivo_entrada, nome_arquivo_saida)



```
